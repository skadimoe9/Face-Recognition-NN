# **FACE RECOGNITION USING CNN**

Face recognition with DeepFace works by using Convolutional Neural Networks (CNNs) to extract unique facial features (embeddings) from images. These embeddings are 128-dimensional vectors that represent the face. The process involves:

1. Face Detection
  
  Locates the face in the image using pre-trained models like MTCNN.
2. Feature Extraction
  
  A CNN (e.g., FaceNet, VGGFace, or ArcFace) converts the face into an embedding.
3. Comparison

  Compares embeddings using a distance metric (e.g., Euclidean) to identify or verify the person.

```python
import os
import random
import shutil
import matplotlib.pyplot as plt
from deepface import DeepFace
import numpy as np
```

```python
# Define paths to your image dataset (entire database folder)
db_path = r"/content/drive/MyDrive/Database/Foto_Resize_CNN_160x160"

# Step 1: Organize dataset (split into training and testing sets)
train_set = []
test_set = []
test_percentage = 0.05  # 5% of the dataset will be used for testing

# Traverse each person's folder in the database
for person in os.listdir(db_path):
    person_folder = os.path.join(db_path, person)
    if os.path.isdir(person_folder):
        person_images = [os.path.join(person_folder, img) for img in os.listdir(person_folder) if img.lower().endswith(('.jpg', '.jpeg', '.png'))]
        random.shuffle(person_images)

        # Split into train and test sets
        split_idx = int(len(person_images) * (1 - test_percentage))
        train_set.extend(person_images[:split_idx])
        test_set.extend(person_images[split_idx:])

print(f"Training set size: {len(train_set)}")
print(f"Testing set size: {len(test_set)}")

# Step 2: Perform face recognition using the test set
for test_image_path in test_set:
    # True label: Extract person's name from the folder structure
    true_label = os.path.basename(os.path.dirname(test_image_path))

    try:
        # Recognize the face in the test image
        result = DeepFace.find(img_path=test_image_path, db_path=db_path)

        # Extract predicted label from the result
        if len(result[0]) > 0:  # Ensure at least one match is found
            predicted_label = os.path.basename(os.path.dirname(result[0]['identity'][0]))  # Extract folder name
        else:
            predicted_label = "Unknown"

    except ValueError as e:
        # Handle the "Face could not be detected" error
        print(f"Error: {e}. Setting predicted label to 'Unknown'")
        predicted_label = "Unknown"

    # Plot the image with the true and predicted labels
    image = plt.imread(test_image_path)
    plt.imshow(image)
    plt.title(f"True: {true_label}\nPred: {predicted_label}")
    plt.axis('off')  # Hide axes
    plt.show()

    print(f"Processed {test_image_path}: True label: {true_label}, Predicted label: {predicted_label}")
```
