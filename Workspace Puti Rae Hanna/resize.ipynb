```python
import os
import cv2

# Define the base input and output folders
base_input_folder = r'D:\Face Recognition\Face-Recognition-NN\Dataset\Foto Terpisah'
base_output_folders = {
    (160, 160): r'D:\Face Recognition\Face-Recognition-NN\Dataset\Foto_Resize_FaceNet_160x160',
}

# List of subfolders for different individuals
subfolders = [
    'Azmira', 'David', 'Dimas', 'Fadhli', 'Fadlin', 'Hafidz',
    'Haidar', 'Hanna', 'Keiko', 'Khansa', 'Mikhael',
    'Puti', 'Raesa', 'Satwika', 'Toni'
]

# Ensure all output folders exist
for resolution, base_output in base_output_folders.items():
    for subfolder in subfolders:
        os.makedirs(os.path.join(base_output, subfolder), exist_ok=True)

# Iterate through each subfolder
for subfolder in subfolders:
    input_folder = os.path.join(base_input_folder, subfolder)

    print(f"Processing folder: {subfolder}")
    for root, dirs, files in os.walk(input_folder):
        for file in files:
            if file.lower().endswith(('.jpg', '.jpeg', '.png')) and not file.startswith('.'):
                try:
                    img_path = os.path.join(root, file)
                    # Read the image as RGB instead of grayscale
                    img = cv2.imread(img_path, cv2.IMREAD_COLOR)  # Read as RGB

                    # Resize the image for each resolution
                    for resolution, base_output in base_output_folders.items():
                        resized_img = cv2.resize(img, resolution)
                        output_file_path = os.path.join(base_output, subfolder, file)
                        cv2.imwrite(output_file_path, resized_img)
                    
                    print(f"{file} successfully resized for all resolutions.")
                except Exception as e:
                    print(f"Error processing {file} in {subfolder}: {e}")
```
